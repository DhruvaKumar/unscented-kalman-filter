```python
import numpy as np
from plotly.offline import init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
nis_radar = np.loadtxt("../build/nis_radar.txt")
nis_laser = np.loadtxt("../build/nis_laser.txt")
```

```python
n = nis_radar.size
n_5 = nis_radar[nis_radar>7.815].size
print(n_5/n)
```

```python
n = nis_laser.size
n_5 = nis_laser[nis_laser>5.991].size
print(n_5/n)
```

```python
# plot nis_radar

t = list(range(0,nis_radar.size))
trace0 = Scatter(
    x=t,
    y=nis_radar,
    name = 'NIS radar'
)

trace1= Scatter(
    x=[t[0], t[-1]],
    y=[7.815, 7.815],
    name = '95 %',
)

data = [trace0, trace1]
fig = Figure(data=data)
iplot(fig, filename= 'nis-radar')
```

```python
# plot nis_laser

t = list(range(0,nis_laser.size))
trace0 = Scatter(
    x=t,
    y=nis_laser,
    name = 'NIS laser'
)

trace1= Scatter(
    x=[t[0], t[-1]],
    y=[5.991, 5.991],
    name = '95 %',
)

data = [trace0, trace1]
fig = Figure(data=data)
iplot(fig, filename= 'nis-laser')
```
